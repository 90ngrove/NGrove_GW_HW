```python
# Dependencies
# Dependencies
import pandas as pd

# load CSV
csv_path = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
code_df =pd.read_csv(csv_path, encoding ='iso-8859-1',low_memory=False)

code_df.head()

# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
new_coders = code_df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
new_coders.head()

# Change "0" to "No" and "1" to "Yes" in response columns
new_coders =new_coders.replace({0.0:"No",1.0:"Yes"})
new_coders.head()
```

```python

# Extract rows corresponding only to people who attended a bootcamp
attended_bootcamp = new_coders.loc [new_coders["AttendedBootcamp"]=="Yes"]
attended_bootcamp.count()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
# Extract rows corresponding only to people who attended a bootcamp


groupedboot= attended_bootcamp[["BootcampName","AttendedBootcamp"]].groupby(['BootcampName']).count()
groupedboot = groupedboot.rename(columns={"BootcampName": "Count"})
groupedboot.head(30)
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
columns = [
    "BootcampName","BootcampRecommend"]

#  Create a new df for "US" with the columns. 
bootcamp_rec_df = attended_bootcamp.loc[attended_bootcamp["BootcampRecommend"] == "Yes",  columns]
bootcamp_recgroup_df= bootcamp_rec_df.groupby(['BootcampName']).count()

bootcamp_recgroup_df.head(30)
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merge_table = pd.merge(groupedboot, bootcamp_recgroup_df, on="BootcampName")

merge_table
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
merge_table["Percent Recommend"] = merge_table['BootcampRecommend'] / \
merge_table['AttendedBootcamp']*100

merge_table = merge_table.sort_values((["Percent Recommend"]),ascending=False)
merge_table["Percent Recommend"] = merge_table["Percent Recommend"].astype(float).map("{:,.2f}%".format)
merge_table
```

```python
# Export to excel and remove index
merge_table.to_excel("bootcamp_recommend.xlsx",index= False)
```
