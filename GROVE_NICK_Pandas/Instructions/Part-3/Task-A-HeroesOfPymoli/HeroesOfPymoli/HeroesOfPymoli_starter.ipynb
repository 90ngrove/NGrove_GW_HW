## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 10, 15, 20, 25, 30, 35, 40, 120]

# Create the names for the four bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
##### Right =False means that the group names are applied up to but not including the right side bound of the bin 
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_names, right=False)
purchase_data.head()
```

```python
purchase_data_group = purchase_data.groupby(['Age Range']).agg({'Purchase ID': 'count'})
purchase_sum =purchase_data_group.sum()
purchase_data_group
```

```python
state_pcts=purchase_data_group.groupby(level=0).apply(lambda x: x/purchase_sum * 100)

merge_purchase = pd.merge(purchase_data_group, state_pcts, on="Age Range")


merge_purchase
```
